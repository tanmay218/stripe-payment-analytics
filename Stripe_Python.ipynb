```python
charges = pd.read_csv("stripe_charges_3000.csv")
balance = pd.read_csv("stripe_balance_3000.csv")

print("Charges shape:", charges.shape)
print("Balance shape:", balance.shape)

print("\nCharges columns:", list(charges.columns))
print("\nBalance columns:", list(balance.columns))
```

```python
charges_clean = charges[[
    'charge_id',
    'amount',
    'currency',
    'description',
    'created_at',
    'card_brand',
    'card_funding',
    'customer_country',
    'product_category',
    'risk_score'
]].copy()

print(charges_clean.head())
print("\nShape:", charges_clean.shape)
```

```python
balance_clean = balance[[
    'txn_id',
    'amount',
    'fee',
    'net',
    'source'
]].copy()

balance_clean.rename(columns={'source': 'charge_id'}, inplace=True)

print(balance_clean.head())
print("\nShape:", balance_clean.shape)
```

```python
print("Charges - Nulls:\n", charges_clean.isnull().sum())
print("\nBalance - Nulls:\n", balance_clean.isnull().sum())

print("\nCharges - Duplicates:", charges_clean.duplicated().sum())
print("Balance - Duplicates:", balance_clean.duplicated().sum())
```

```python
charges_clean.to_csv("charges_clean.csv", index=False)
balance_clean.to_csv("balance_clean.csv", index=False)

print("Saved successfully!")
```
